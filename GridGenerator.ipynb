```python
from ssgg.ssgg import grid_generator, combine_mesh, plot_grid, write_mesh
import pandas as pd
```

## Radial Grid

```python
Radius = 1 # @param {"type":"slider","min":1,"max":5,"step":1}
Radius_Expansion_Ratio = 0.6 # @param {"type":"slider","min":0.1,"max":2,"step":0.1}
Number_Radial_Cells = 3 # @param {"type":"slider","min":2,"max":50,"step":1}
```

## Sector 1 Angular Grid

```python
Initial_Angle_1 = 0 # @param {"type":"number","placeholder":"0"}
Sector_Angle_1 = 45 # @param {"type":"integer","placeholder":"45"}
Number_Angular_Cells_1 = 3 # @param {"type":"slider","min":2,"max":50,"step":1}
Angle_Expansion_Ratio_1 = 0.6 # @param {"type":"number","placeholder":"0.5"}
```

## Sector 2 Angular Grid

```python
Initial_Angle_2 = Initial_Angle_1 + Sector_Angle_1
Sector_Angle_2 = 45 # @param {"type":"integer","placeholder":"45"}
Number_Angular_Cells_2 = 5 # @param {"type":"slider","min":2,"max":50,"step":1}
Angle_Expansion_Ratio_2 = 1/Angle_Expansion_Ratio_1

mesh1 = grid_generator(radius=Radius, sector_angle=Sector_Angle_1,
                    number_of_radial_cells=Number_Radial_Cells, number_of_angular_cells=Number_Angular_Cells_1,
                    radial_expansion_ratio=Radius_Expansion_Ratio, angular_expansion_ratio=Angle_Expansion_Ratio_1,
                    radius_0=0, sector_angle_0=Initial_Angle_1)

mesh2 = grid_generator(radius=Radius, sector_angle=Sector_Angle_2,
                    number_of_radial_cells=Number_Radial_Cells, number_of_angular_cells=Number_Angular_Cells_2,
                    radial_expansion_ratio=Radius_Expansion_Ratio, angular_expansion_ratio=Angle_Expansion_Ratio_2,
                    radius_0=0, sector_angle_0=Initial_Angle_2)

mesh = combine_mesh(mesh1, mesh2)
```

# Plotting Controls

```python
Node_Color = "Blue" # @param ["Blue","Red","Black"]
Edge_Color = "Black" # @param ["Black","Red","Blue"]
Edge_Linewidth = 0.7 # @param {"type":"slider","min":0.1,"max":2,"step":0.1}
Show_Axes = False # @param {type:"boolean"}
Show_Node_Numbers = True # @param {type:"boolean"}
Show_Element_Numbers = True # @param {type:"boolean"}
Save_Figure = True # @param {type:"boolean"}

plotting_inputs = {
    "node_color":Node_Color,
    "edge_color":Edge_Color,
    "edge_linewidth":Edge_Linewidth,
    "show_axes":Show_Axes,
    "show_node_numbers":Show_Node_Numbers,
    "show_element_numbers":Show_Element_Numbers,
    "save_figure":Save_Figure
}
plot_grid(mesh, plotting_inputs)
```

# Save Grid Data

```python
Filename_Suffix = "_R_1_Angle_90" # @param {"type":"string","placeholder":"_R_1_Angle_90"}
write_mesh(mesh, Filename_Suffix)
```
